```sql
CREATE OR REPLACE TABLE nestle_dev_gold.bi_core.dim_product AS
SELECT DISTINCT
  src.product_id,
  COUNT(*) as total_transactions,
  SUM(src.amount) as total_revenue,
  MIN(src.dbt_valid_from) as first_transaction_date,
  MAX(src.dbt_valid_from) as last_transaction_date
FROM nestle_dev_silver.core.sales_transactions_scd2 src
WHERE src.dbt_is_current = TRUE
GROUP BY src.product_id;

SELECT COUNT(*) as product_count FROM nestle_dev_gold.bi_core.dim_product;
```

```sql
CREATE TABLE IF NOT EXISTS nestle_dev_gold.bi_core.dim_region AS
SELECT DISTINCT
  src.region,
  COUNT(*) as transaction_count,
  COUNT(DISTINCT src.product_id) as unique_products,
  SUM(src.amount) as total_amount
FROM nestle_dev_silver.core.sales_transactions_scd2 src
WHERE src.dbt_is_current = TRUE
GROUP BY src.region;

SELECT COUNT(*) as region_count FROM nestle_dev_gold.bi_core.dim_region;
```

```sql
CREATE TABLE IF NOT EXISTS nestle_dev_gold.bi_core.dim_channel AS
SELECT DISTINCT
  src.channel,
  COUNT(*) as transaction_count,
  SUM(src.amount) as total_amount,
  AVG(src.amount) as avg_transaction_value
FROM nestle_dev_silver.core.sales_transactions_scd2 src
WHERE src.dbt_is_current = TRUE
GROUP BY src.channel;

SELECT COUNT(*) as channel_count FROM nestle_dev_gold.bi_core.dim_channel;
```

```sql
SELECT 
  (SELECT COUNT(*) FROM nestle_dev_gold.bi_core.dim_product) as product_dims,
  (SELECT COUNT(*) FROM nestle_dev_gold.bi_core.dim_region) as region_dims,
  (SELECT COUNT(*) FROM nestle_dev_gold.bi_core.dim_channel) as channel_dims;
```
