```python
%sql
CREATE OR REPLACE TABLE nestle_dev_bronze.sharepoint.regional_sales_targets AS
SELECT DISTINCT *
FROM nestle_dev_bronze.sharepoint.regional_sales_targets;
```

```python
%sql
SELECT 
  COUNT(*) AS total_rows,
  COUNT(DISTINCT concat_ws('||', product_id, region, period, cast(target_amount as string), row_hash, cast(ingestion_timestamp as string), source_system_name, source_dataset_name, run_id, pipeline_name)) AS distinct_full_rows,
  COUNT(*) - COUNT(DISTINCT concat_ws('||', product_id, region, period, cast(target_amount as string), row_hash, cast(ingestion_timestamp as string), source_system_name, source_dataset_name, run_id, pipeline_name)) AS remaining_exact_duplicates
FROM nestle_dev_bronze.sharepoint.regional_sales_targets;
```
